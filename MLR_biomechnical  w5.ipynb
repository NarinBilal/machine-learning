```python
data=data.rename(columns={"pelvic_tilt numeric": "pelvic_tilt"}) # rename the column name
```

```python
data.head()
```

```python
data['class']=[1 if each=="Abnormal" else 0 for each in data['class']]
```

```python
x=data[['pelvic_incidence','pelvic_tilt','lumbar_lordosis_angle','sacral_slope','pelvic_radius','degree_spondylolisthesis']]
y=data['class']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()  
mlr.fit(x_train, y_train)
```

```python
print("Intercept: ",mlr.intercept_)
print("coefficients: ")
list(zip(x,mlr.coef_))
```

```python
#prediction of test set
y_pred_mlr=mlr.predict(x_test)
#predicted values
print("prediction for test set: {}".format(y_pred_mlr))
```

```python
#actual value and the predicted value
mlr_diff = pd.DataFrame({'Actual value': y_test, 'predicted value': y_pred_mlr})
mlr_diff.head()
```

```python
#model Evaluation 
from sklearn import metrics
meanSqErr = metrics.mean_squared_error(y_test,y_pred_mlr)
print('R squared: {Mean Square Error: }',meanSqErr)
```

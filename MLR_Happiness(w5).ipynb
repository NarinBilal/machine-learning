```python
data=data.rename(columns={"Overall rank": "Overall_rank",
                          "Country or region": "Country_or_region",
                          "GDP per capita": "GDP_per_capita",
                          "Social support": "Social_support",
                          "Healthy life expectancy": "Healthy_life_expectancy",
                          "Freedom to make life choices": "Freedom_to_make_life_choices",
                          "Perceptions of corruption": "Perceptions_of_corruption"}) # rename the column name
```

```python
data.head()
```

```python
x=data[['Overall_rank','Year','GDP_per_capita','Social_support','Healthy_life_expectancy','Freedom_to_make_life_choices','Generosity','Perceptions_of_corruption']]
y=data['Score']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
data
```

```python
x.info()
```

```python
data['Score'].unique()
```

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()  
mlr.fit(x_train, y_train)
```

```python
print("Intercept: ",mlr.intercept_)
print("coefficients: ")
list(zip(x,mlr.coef_))
```

```python
#prediction of test set
y_pred_mlr=mlr.predict(x_test)
#predicted values
print("prediction for test set: {}".format(y_pred_mlr))
```

```python
#actual value and the predicted value
mlr_diff = pd.DataFrame({'Actual value': y_test, 'predicted value': y_pred_mlr})
mlr_diff.head()
```

```python
#model Evaluation 
from sklearn import metrics
meanSqErr = metrics.mean_squared_error(y_test,y_pred_mlr)
print('R squared: {Mean Square Error: }',meanSqErr)
```
